```python
import torch
from transformers import BertJapaneseTokenizer, BertForMaskedLM
tokenizer = BertJapaneseTokenizer.from_pretrained('cl-tohoku/bert-base-japanese-whole-word-masking')
```

```python
text = '父の父は、祖父。'
tokenized_text = tokenizer.tokenize(text)
print(tokenized_text)
```

```python
masked_index = 2
tokenized_text[masked_index] = '[MASK]'
print(tokenized_text)
```

```python
indexed_tokens = tokenizer.convert_tokens_to_ids(tokenized_text)
tokens_tensor = torch.tensor([indexed_tokens])
print(tokens_tensor)
```

```python
# Load pre-trained model
model_mask = BertForMaskedLM.from_pretrained('cl-tohoku/bert-base-japanese-whole-word-masking')
model_mask.eval()
```

```python
# Predict
with torch.no_grad():
    outputs = model_mask(tokens_tensor)
    predictions = outputs[0][0, masked_index].topk(5) # 予測結果の上位５件を抽出

# Show results
for i, index_t in enumerate(predictions.indices):
    index = index_t.item()
    token = tokenizer.convert_ids_to_tokens([index])[0]
    print(i, token)
```

```python
from transformers import BertJapaneseTokenizer, BertModel
tokenizer = BertJapaneseTokenizer.from_pretrained('cl-tohoku/bert-base-japanese-whole-word-masking')
model_bert = BertModel.from_pretrained('cl-tohoku/bert-base-japanese-whole-word-masking')
model_bert.eval()
```

```python
import numpy as np

def calc_embadding(text):
    bert_tokens = tokenizer.tokenize(text)
    ids = tokenizer.convert_tokens_to_ids(["[CLS]"] + bert_tokens[:126] + ["[SEP]"])
    tokens_tensor = torch.tensor(ids).reshape(1, -1)

    with torch.no_grad():
        output = model_bert(tokens_tensor)

    return output[1].numpy()
```

```python
text = '私はからあげです'
calc_embadding(text)
```

```python
import pandas as pd

data = pd.read_csv('articles_list.csv')
data.head()
```

```python
title_list = []

embedding_list = []

for title in data['Article TITLE']:
    title_list.append(title)
    embedding_list.append(calc_embadding(title))
```

```python
with open('metadata_bert.tsv', 'w', encoding='utf-8') as f:
    for title in title_list:
        f.write(title[:-18] + '\n') # -18の数字は記事タイトルからブログタイトルを削除するための処理

with open('vector_bert.tsv', 'w', encoding='utf-8') as f:
    for embedding in embedding_list:
        for vec in embedding[0]:
            f.write(str(vec) + '\t')
        f.write('\n')
```
